```python

# We want to extract a sample of 10000 lines to make sure the data can be used properly. 

n_samples = min(10000, len(bigbase))
bigbase_sample = bigbase.sample(n=n_samples, random_state=42)  
bigbase_sample.to_csv("Montgomery_sample.csv", index=False)

print("Shape original :", bigbase_sample.shape)
print(bigbase_sample.head(100))
```

```python

# Let's now clean the database to make sure there are no missing values for the coordonates of the car crash 
bigbase_sample_clean = bigbase_sample.dropna(subset=["Latitude", "Longitude"])

print(bigbase_sample_clean.shape)
bigbase_sample_clean.head()

columns = [
    "Non-Motorist Substance Abuse",
    "Surface Condition",
    "Weather",
    "Collision Type",
    "Injury Severity",
    "Vehicle Movement","Vehicle Damage Extent","Driver Distracted By","Driverless Vehicle"
]

# Afficher les 5 premières lignes uniquement de ces colonnes
bigbase_sample_clean[columns].head()
```

```python
for col in bigbase_sample_clean.columns:
    print(f"Colonne : {col}")
    
    if pd.api.types.is_numeric_dtype(bigbase_sample_clean[col]):
        print("  Min :", bigbase_sample_clean[col].min())
        print("  Max :", bigbase_sample_clean[col].max())
    print("-" * 50)
```

```python
# Liste des colonnes que tu veux explorer
colonnes_a_explorer = [
    "Non-Motorist Substance Abuse",
    "Surface Condition",
    "Weather",
    "Collision Type",
    "Injury Severity",
    "Vehicle Movement"
]

# Afficher les valeurs uniques pour chaque colonne
for col in colonnes_a_explorer:
    valeurs_uniques = bigbase_sample_clean[col].dropna().unique()
    print(f"Colonne : {col}")
    print(f"Valeurs uniques ({len(valeurs_uniques)}): {valeurs_uniques}")
    print("-" * 50)
```
